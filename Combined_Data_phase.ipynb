```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os, re
from sklearn.utils import shuffle
```

```python
train_df = pd.read_csv('training_dataset_cubic_phase.csv', sep='\t')
test_df = pd.read_csv("phase_test.csv", sep='\t')
```

```python
train_df = shuffle(train_df)
```

```python
al_sample = shuffle(train_df[train_df['names'] == 'Aluminium']).iloc[:100]
al_sample = al_sample[['peak_1', 'peak_2', 'peak_3', 'peak_4', 'peak_5', 'peak_6', "peak_7", "peak_8", 'names']]

fe_sample = shuffle(train_df[train_df['names'] == 'Iron']).iloc[:100]
fe_sample = fe_sample[['peak_1', 'peak_2', 'peak_3', 'peak_4', 'peak_5', 'peak_6', "peak_7", "peak_8", 'names']]
```

```python
peaks = []
codes = []
```

```python
al_sample
```

```python
# getting the sum of two different materials
for i in range(len(al_sample)):
    for j in range(len(fe_sample)):
        peaks.append(al_sample.iloc[i][:8].values + fe_sample.iloc[j][:8].values)
        codes.append(0)
        
for i in range(len(al_sample)):
    for j in range(len(al_sample)):
        peaks.append(al_sample.iloc[i][:8].values + al_sample.iloc[j][:8].values)
        codes.append(1)
        
for i in range(len(fe_sample)):
    for j in range(len(fe_sample)):
        peaks.append(fe_sample.iloc[i][:8].values + fe_sample.iloc[j][:8].values)
        codes.append(2)
        
```

```python
data = pd.DataFrame(peaks, columns=["peak_1", "peak_2", "peak_3", "peak_4","peak_5", "peak_6", "peak_7", "peak_8"])
data['codes'] = codes
```

```python
data = shuffle(data)
data = data.reset_index(drop=True)
data
```

```python
data.to_csv("combined_phase.csv", sep='\t', index=False)
```
